```
import math
import random
import matplotlib.pyplot as plt
# create random data
no_of_balls = 10
x = [random.gauss(0, 1) for i in range(no_of_balls)]
y = [random.gauss(0, 1) for i in range(no_of_balls)]
colors = [4 for i in range(no_of_balls)]
areas = [math.pi * 5**2 for i in range(no_of_balls)]
# draw the plot
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
plt.axis([-10.0, 10.0, -10.0, 10.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
# draw the turned plot
plt.figure()
plt.scatter(y, x, s=areas, c=colors, alpha=0.85)
plt.axis([-10.0, 10.0, -10.0, 10.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

#create tranpose unitary matrix
a = random.uniform(0,1)
b = math.sqrt(1 - a**2)
thetainitial = math.atan(b/a)
thetafinal = thetainitial+(math.pi/2)
c = math.cos(thetafinal)
d = math.sin(thetafinal)
#create tranposed data
#[ab] [cd] have to be orthogonal
'''
a b  *  x  =  ax + by  =  x2
c d  *  y  =  cx + dy  =  y2
'''
ax = [a*i for i in x]
by = [b*i for i in y]
cx = [c*i for i in x]
dy = [d*i for i in y]
zipped_lists1 = zip(ax, by)
x2 = [x + y for (x, y) in zipped_lists1]
zipped_lists2 = zip(cx, dy)
y2 = [x + y for (x, y) in zipped_lists2]
colors2 = [2 for i in range(no_of_balls)]
# draw the plot
plt.figure()
plt.scatter(x2, y2, s=areas, c=colors2, alpha=0.85)
plt.axis([-10.0, 10.0, -10.0, 10.0])
plt.xlabel("X2")
plt.ylabel("Y2")
plt.show()
```
